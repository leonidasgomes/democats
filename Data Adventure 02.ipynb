```python
import urllib.request
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('ggplot')
%matplotlib inline
```

### 2 - Baixando a Base de Dados

```python
print('Baixando os arquivos...')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2013.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2013.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2014.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2014.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2015.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2015.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2016.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2016.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2017.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2017.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2018.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2018.csv')

url = 'http://ftp.dadosabertos.ans.gov.br/FTP/PDA/demandas_dos_consumidores_nip/PDA_NIP_2019.csv'
urllib.request.urlretrieve(url, 'PDA_NIP_2019.csv')

```

### 3- Lendo a Base de Dados
Dica:Sempre verifique o separador do csv e o enconding do idioma, isso é necessario devido a acentuação do português.

```python
df2013=pd.read_csv('PDA_NIP_2013.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2014=pd.read_csv('PDA_NIP_2014.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2015=pd.read_csv('PDA_NIP_2015.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2016=pd.read_csv('PDA_NIP_2016.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2017=pd.read_csv('PDA_NIP_2017.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2018=pd.read_csv('PDA_NIP_2018.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')
df2019=pd.read_csv('PDA_NIP_2019.csv', sep=';',encoding='latin1',low_memory=False, decimal=',')

df=pd.concat([df2013,df2014,df2015,df2016,df2017,df2018,df2019])
```

### 4- Verificando as Colunas do Arquivo

```python
df.columns
```

### 4- Verificando as Colunas e Tipo de dados

```python
df.info()
```

```python
df['NUMERO_DA_DEMANDA'].nunique()
df = df[(df['MES_DE_REFERENCIA']<=6)]
```

```python
dfm =  df[['ANO_DE_REFERENCIA','NUMERO_DA_DEMANDA']].groupby(['ANO_DE_REFERENCIA'])['NUMERO_DA_DEMANDA'] \
                             .nunique() \
                             .reset_index(name='Demandas') 
dfano=  df[['ANO_DE_REFERENCIA','NUMERO_DA_DEMANDA']].groupby(['ANO_DE_REFERENCIA'])['NUMERO_DA_DEMANDA'] \
                             .nunique() \
                             .reset_index(name='Demandas') \

      


      
```

```python
dfm.plot(kind = 'bar',x='ANO_DE_REFERENCIA',y='Demandas',rot=5,fontsize=8,  figsize=(10,5),title='NIP - Volume de Demandas Por Ano - Gráfico de Barras')
dfano.plot(kind = 'line',x='ANO_DE_REFERENCIA',y='Demandas',rot=5,fontsize=8,  figsize=(10,5),title='NIP - Volume de Demandas Por Ano - Gráfico de Linha')
 
```
